```python
#a. 
data12 = pd.read_sas("/DEMO_G.XPT")
data34 = pd.read_sas("/DEMO_H.XPT")
data56 = pd.read_sas("/DEMO_I.XPT")
data78 = pd.read_sas("/DEMO_J.XPT")

#Add an id to each cohort 
data12["Cohort_ID"] = [1 for i in range(data12.shape[0])]
data34["Cohort_ID"] = [2 for i in range(data34.shape[0])]
data56["Cohort_ID"] = [3 for i in range(data56.shape[0])]
data78["Cohort_ID"] = [4 for i in range(data78.shape[0])]

data1 = data12[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
                   "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR", "Cohort_ID"]] 
data2 = data34[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
                   "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR", "Cohort_ID"]] 
data3 = data56[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
                   "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR", "Cohort_ID"]] 
data4 = data78[["SEQN", "RIDAGEYR", "RIDRETH3", "DMDEDUC2", "DMDMARTL", 
                   "RIDSTATR", "SDMVPSU", "SDMVSTRA", "WTMEC2YR", "WTINT2YR", "Cohort_ID"]]

#append 
frame = [data1, data2, data3, data4]
data_clean = pd.concat(frame)

#rename the columns 
data_clean = data_clean.rename(columns = {"SEQN": "respondent sequence number", "RIDAGEYR": 
                                                "age in years at screening", 
                                                "RIDRETH3": "race/hispanic origin w/ nh asian", 
                                                "DMDEDUC2": "education level - adults 20+", 
                                                "DMDMARTL": "marital status", 
                                                "RIDSTATR": "interview/examination status", 
                                                "SDMVPSU": "masked variance pseudo-PSU", 
                                                "SDMVSTRA": "masked variance pseudo-stratum", 
                                                "WTMEC2YR": "full sample 2 year mec exam weight", 
                                                "WTINT2YR": "full sample 2 year interview weight"})

#replace NaN with -1
data_clean = data_clean.fillna(-1)

#convert the data structure type in each column
convert_data = {}
for i in data_clean.columns:
    if i != "WTMEC2YR" and i != "WTINT2YR":
        convert_data[i] = int

data_clean = data_clean.round(2) #to round WTMEC2YR and WTINT2YR to two decimal digits 
data_clean = data_clean.astype(convert_data)

#serialize the data with pickle
serial_data = pickle.dumps(data_clean)
pickle.loads(serial_data)
```
